```python
# Step 1: Install Hugging Face Transformers and SentencePiece
!pip install transformers sentencepiece

# Step 2: Import necessary libraries
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer

# Step 3: Initialize the tokenizer and model for M2M-100 (1.2B parameters)
model_name = "facebook/m2m100_1.2B"
tokenizer = M2M100Tokenizer.from_pretrained(model_name)
model = M2M100ForConditionalGeneration.from_pretrained(model_name)

```

```python
# Step 4: Define source and target languages
source_lang = "fr"  # Example: French
target_lang = "de"  # Example: German
tokenizer.src_lang = source_lang

# Step 5: Prepare input text and tokenize
text = "Bonjour tout le monde! it is good"  # Text in source language (French)
encoded_text = tokenizer(text, return_tensors="pt")
```

```python

# Step 6: Generate translation
generated_tokens = model.generate(**encoded_text, forced_bos_token_id=tokenizer.get_lang_id(target_lang))

# Step 7: Decode and print the output
translation = tokenizer.decode(generated_tokens[0], skip_special_tokens=True)
print(f"Translation ({source_lang} to {target_lang}):", translation)
```
